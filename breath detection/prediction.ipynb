```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import librosa
import librosa.display
import IPython.display as ipd
import tensorflow as tf
import matplotlib.pyplot as plt


apath = '/home/egor/safeVision/code/Snoring Dataset/0/0_1.wav'

MFCC_NUM =  28
SAMPLING_RATE = 44100 
MFCC_MAX_LEN = 128 


def wav2mfcc(wave, max_len=MFCC_MAX_LEN):
    mfcc = librosa.feature.mfcc(y=wave, n_mfcc=MFCC_NUM, sr=SAMPLING_RATE)
    if (max_len > mfcc.shape[1]):
        pad_width = max_len - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad_width)), mode='constant')

    else:
        mfcc = mfcc[:, :max_len]
    
    return mfcc


    
def predict_audio_class(apath):
    audio, _ = librosa.load(apath, sr=SAMPLING_RATE)
    spectrogram = np.expand_dims(wav2mfcc(audio), axis=-1)
    spectrogram = np.expand_dims(spectrogram, axis=0)
    model = tf.keras.models.load_model('/home/egor/safeVision/SafeVision/breath detection/models/model_snoring_and_breathing_10e_28_128.keras')
    prediction = model.predict(spectrogram)
    predicted_class = np.argmax(prediction, axis=-1)
    
    print(f"Prediction: {predicted_class[0]}")
    return predicted_class[0]
```

```python
AUDIO_PATH = '/home/egor/safeVision/code/Snoring Dataset/1'  # Путь к папке с аудио файлами

# Пример обработки нескольких файлов
for i in range(100):
    file_path = os.path.join(AUDIO_PATH, f'1_{i}.wav')
    
    if os.path.exists(file_path):
        predicted_class = predict_audio_class(file_path)
        # if predicted_class > 0.5:
        #     print('snor or breath')
        # else:
        #     print('no breath')
        print(f"File {file_path}: Predicted class = {predicted_class}")
```
